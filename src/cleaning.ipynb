```python
# Camel-Casing my cleaning (as listed in instructions!)
df = df.rename(columns={
    "Passengers_In": "PassengersIn",
    "Freight_In_(tonnes)": "FreightIn",
    "Mail_In_(tonnes)": "MailIn",
    "Passengers_Out": "PassengersOut",
    "Freight_Out_(tonnes)": "FreightOut",
    "Mail_Out_(tonnes)": "MailOut",
    "Passengers_Total": "PassengersTotal",
    "Freight_Total_(tonnes)": "FreightTotal",
    "Mail_Total_(tonnes)": "MailTotal",
    "Month_num": "MonthNum",
})
```

```python
df["Quarter"] = ((df["MonthNum"] - 1) // 3) + 1
df["MonthYear"] = pd.to_datetime(df["Year"].astype(str) + "-" + df["MonthNum"].astype(str) + "-01")

country_continent_mapping = {
    "New Zealand": "Oceania",
    "Bahrain": "Asia",
    "India": "Asia",
    "Germany": "Europe",
    "UK": "Europe",
    "Oman": "Asia",
    "Italy": "Europe",
    "Singapore": "Asia",
    "United Arab Emirates": "Asia",
    "Thailand": "Asia",
    "Hong Kong": "Asia",
    "Solomon Islands": "Oceania",
    "USA": "North America",
    "Malaysia": "Asia",
    "Philippines": "Asia",
    "Fiji": "Oceania",
    "New Caledonia": "Oceania",
    "Papua New Guinea": "Oceania",
    "Vanuatu": "Oceania",
    "Japan": "Asia",
    "Canada": "North America",
    "Brunei": "Asia",
    "Indonesia": "Asia",
    "Netherlands": "Europe",
    "Greece": "Europe",
    "China": "Asia",
    "Yugoslavia": "Europe",
    "Sri Lanka": "Asia",
    "Cyprus": "Asia",
    "Mauritius": "Africa",
    "Nauru": "Oceania",
    "Tahiti": "Oceania",
    "Zimbabwe": "Africa",
    "South Africa": "Africa",
    "Western Samoa": "Oceania",
    "France": "Europe",
    "Denmark": "Europe",
    "Cook Islands": "Oceania",
    "American Samoa": "Oceania",
    "Austria": "Europe",
    "Argentina": "South America",
    "Guam": "Oceania",
    "Malta": "Europe"
}
df["Continent"] = df["Country"].map(country_continent_mapping)
```
